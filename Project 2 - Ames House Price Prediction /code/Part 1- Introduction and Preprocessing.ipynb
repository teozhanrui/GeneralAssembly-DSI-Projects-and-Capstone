# Introduction

This machine learning model is targeted at realtors to help them assess the market value of properties. The pandemic due to COVID 19 has changed many aspects of our lives and most importantly, the way we work. The real estate sector has been tremendously disrupted as contact between clients and realtors is reduced. Virtual viewings are now common and widely accepted before the clients take a step forward to express further interest. 

In this age of digitalization where efficiency is priced, clients often expect realtors to do back of the envelope estimations of market value on the fly based on experience and market transactions. A detailed assessment by a valuer takes some time and is often done only when the buyer is ready to commit. Negotiation with the seller is then largely based on the estimation. When the actual valuation does not match the initial estimation, expectations are not met and loan approval may be affected, leading to potential loss of deal. 

In this project, only **linear regression** with regularizations using ridge, lasso and elastic net are used.

## Problem Statement

1) Investigate how well the linear regression predicts past prices. Recommend applications and limitations of model to future predictions.

2) Determine the most important fixed features in determining the sale price of a property.

3) Identify the most important features that can be improved by renovation to drive up the property price. Similarly, identify the most undesirable features that are detrimental to property price.



## Executive Summary

1) The model predicts well for properties of prices up to USD 350,000. Beyond that, the model often under-estimates the price. 

2) Fixed features: Lot area, gross living area and location determine the sale price the most. 

Stone Brook and Northridge Villa are the best neighborhoods. Those in Department of Transport and Rail Road fetch the lowest prices 

Properties near off-site features like parks and greenbelt improve prices while those near the railway are undesirable. Townhouses and properties with high lot frontage are negatively impacted.

3) Renovation: Improvement to kitchen quality can help lift prices. Excessive garage area hurt prices.

**Renaming of Columns**

```python
# custom function to rename columns to lower case and replace whitespaces with underscore.
def rename_cols(df):
    old_cols = df.columns
    new_cols = [x.lower().replace(' ','_') for x in old_cols]
    #create dictionary with old column label as key and new label as value
    col_dict = {old:new for old, new in zip(old_cols,new_cols)}
    df.rename(columns=col_dict, inplace=True)
    return df
```

# Feature Engineering

## Combining Features

### Variables Related to Age

```python
year_cols = [col for col in ames.columns if ('year' in col) or ('yr' in col)]
```

```python
ames[year_cols].head()
```

Observations: I will take the difference between the year of sale and year built to reflect the age of the property at the time of transaction. The year remodelled can be significantly later than year built for old property. I believe a client would judge largely based on the condition of the property over the year remodelled. However very old houses may attract a niche market due to their uniqueness and heritage, provided they are of decent condition. Hence I will base my model on the year built and ignore the year remodelled.

```python
# new column for age at time of sale
ames['age'] = ames['yr_sold']-ames['year_built']
```

```python
ames.drop(columns=['year_built','year_remod/add','yr_sold'], inplace=True)
```

### Variables Related to Area

```python
area_cols = [x for x in ames.columns if 'area' in x or 'sf' in x or 'porch' in x]
```

```python
ames[area_cols].head()
```

Observations: Most of the areas are subsets of a larger area, such as basement and gross living area. Taking all areas into account will introduce much granularity but can lead to overfitting. 

I will select only the main functional areas and drop the rest. In addition, the garage area is sufficiently representative, so the number of cars in garage is redundant and subjective. All other features will be combined as exterior built up area ('ext_blt_area').

```python
# new feature that takes sum of all exterior areas of built up features
ames['ext_blt_area'] = ames['mas_vnr_area']+ames['wood_deck_sf']+ames['open_porch_sf']+\
                       ames['enclosed_porch']+ames['3ssn_porch']+ames['screen_porch']
```

Observations: All these variables are ordinal and they use very similar manner of rankings and labels. I will replace them with number and do a simple average.
```python
# custom function for boxplots for categorical variables
def subplot_boxplots(dataframe, list_of_columns):
    nrows = int(np.ceil(len(list_of_columns)/2)) 
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize=(16, nrows*3.5)) 
    # Enter figsize as tuple (m,n)
    ax = ax.ravel() 
    for i, column in enumerate(list_of_columns): 
        sns.boxplot(ax=ax[i], x='saleprice', y=column, data=dataframe)
        # ax[i].set_title(column, fontsize=12)
        ax[i].set_xlabel('sale price (USD)',fontsize=13)
        ax[i].set_ylabel(column, fontsize=13)
        ax[i].tick_params('x', labelsize=12)
        ax[i].tick_params('y', labelsize=12)
    plt.tight_layout()
```

```python
# so as to produce box plots for these variables
ames[['overall_qual','overall_cond']]=ames[['overall_qual','overall_cond']].astype('category')
```

```python
# custom function for scatter plots for continous variables against sale price
def subplot_scatplots(dataframe, list_of_columns):
    nrows = int(np.ceil(len(list_of_columns)/2)) 
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize=(16, nrows*3.5)) 
    # Enter figsize as tuple (m,n)
    ax = ax.ravel() 
    for i, column in enumerate(list_of_columns): 
        sns.regplot(ax=ax[i], x=column, y='saleprice', data=dataframe, line_kws={'color':'green'})
        # ax[i].set_title(column, fontsize=12)
        ax[i].set_xlabel(column,fontsize=13)
        ax[i].set_ylabel('sale price (USD)', fontsize=13)
        ax[i].tick_params('x', labelsize=12)
        ax[i].tick_params('y', labelsize=12)
    plt.tight_layout()
```
